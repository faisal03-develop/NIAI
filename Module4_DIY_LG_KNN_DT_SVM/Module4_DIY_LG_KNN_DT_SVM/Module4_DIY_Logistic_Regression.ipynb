## What is Logistic Regression?

**Logistic Regression** is a statistical method used for binary classification problems. Despite containing "regression" in its name, it's actually a **classification algorithm** that predicts the probability of an instance belonging to a particular class.

### Key Concepts:

#### The Sigmoid Function
The core of logistic regression is the **sigmoid (logistic) function**:

**Formula**: σ(z) = 1 / (1 + e^(-z))

Where: z = β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ

#### Why Sigmoid?
- **Output range**: Always between 0 and 1 (perfect for probabilities)
- **S-shaped curve**: Smooth transition from 0 to 1
- **Interpretable**: Output represents probability of belonging to class 1

#### Decision Making:
- **If P(Zombie) > 0.5**: Classify as Zombie (Class 1)
- **If P(Zombie) ≤ 0.5**: Classify as Human (Class 0)

### Advantages for Binary Classification:
1. **Probabilistic Output**: Know confidence in predictions
2. **Fast Prediction**: Quick classification once trained
3. **Interpretable**: Understand which features matter most
4. **Memory Efficient**: Only stores learned weights
5. **No Hyperparameters**: Works well with default settings
