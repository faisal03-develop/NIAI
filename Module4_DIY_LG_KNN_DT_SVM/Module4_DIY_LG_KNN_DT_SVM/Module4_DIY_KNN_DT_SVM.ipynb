### What is K-Nearest Neighbors (KNN)?

KNN is a **lazy learning algorithm** that makes predictions based on the similarity of data points. It's called "lazy" because it doesn't build an explicit model during training - instead, it stores all training data and makes decisions at prediction time.

### How KNN Works (Step by Step):

1. **Store all training data** (images and their labels)
2. **For each test image**:
   - Calculate distance to ALL training images
   - Find the K closest training images (neighbors)
   - Take a "vote" from these K neighbors
   - Assign the most common class label

### Why KNN for Image Classification?

**Intuition**: Similar-looking clothes should be classified the same way!
- A dress image should be more similar to other dress images than to pants
- Shoes have distinct visual features that differ from shirts
- KNN captures this similarity through distance measurements

### Visual Example:
```

- Neighbor 3: Pants (distance: 0.18)
                ↓

                ↓
Prediction: "Dress" ✅
```

## Distance Functions in KNN

### Why Distance Matters?
Distance functions determine how we measure "similarity" between images. Different distance metrics can lead to different classification results!

### 1. Euclidean Distance
**Formula**: √[(x₁-y₁)² + (x₂-y₂)² + ... + (xₙ-yₙ)²]

**Intuition**: 
- "Straight-line" distance in multi-dimensional space
- Like measuring the direct distance between two points on a map
- Sensitive to large differences in individual pixel values

**For Images**:
- Each pixel is a dimension in our feature space
- For 28×28 images = 784 dimensions!
- Measures overall pixel-wise similarity

**Example**: If two dress images have similar overall brightness and color patterns, Euclidean distance will be small.

### 2. Manhattan Distance  
**Formula**: |x₁-y₁| + |x₂-y₂| + ... + |xₙ-yₙ|

**Intuition**:
- "City block" distance - like walking through city streets
- Sum of absolute differences
- Less sensitive to outliers than Euclidean distance

**For Images**:
- Adds up absolute pixel differences
- More robust to extreme pixel values
- Good when images have similar structure but different intensities

**Example**: Manhattan distance might work better when comparing a bright white shirt to a dark black shirt of the same style.

## Evaluation Metrics Explained



### 1. Accuracy
**Formula**: (Correct Predictions) / (Total Predictions)

**What it tells us**:
- Overall correctness of our model
- Simple and intuitive metric

**Example**: 
- Total test images: 1000
- Correct predictions: 850
- Accuracy = 850/1000 = 85%

**Limitation**: Can be misleading with imbalanced datasets!

### 2. Precision 
**Formula**: Average of precision for each class
- Precision for class = True Positives / (True Positives + False Positives)

**What it tells us**:
- "When I predict 'dress', how often am I right?"
- Measures false positive rate
- Important when wrong positive predictions are costly

**Example for "Dress" class**:
- Predicted 100 images as "dress"
- 85 were actually dresses, 15 were not
- Precision = 85/100 = 85%

**Why Macro-Average?**: Treats all classes equally, regardless of their frequency

### 3. Recall 
**Formula**: Average of recall for each class
- Recall for class = True Positives / (True Positives + False Negatives)

**What it tells us**:
- "Of all actual dresses, how many did I find?"
- Measures false negative rate
- Important when missing positive cases is costly

**Example for "Dress" class**:
- 120 actual dress images in test set
- Model correctly identified 85 of them
- Recall = 85/120 = 70.8%

## What is a Decision Tree?

**Decision Trees** are a non-parametric supervised learning method used for both classification and regression. They create a model that predicts the target value by learning simple decision rules inferred from the data features.

### Key Concepts:

#### Tree Structure:
- **Root Node**: The topmost node representing the entire dataset
- **Internal Nodes**: Represent feature tests (e.g., "Is pixel[100] > 0.5?")
- **Leaf Nodes**: Represent class predictions (e.g., "Dress", "Pants")
- **Branches**: Connect nodes and represent outcomes of tests

#### How Decision Trees Work:
1. **Start at Root**: Begin with the entire training dataset
2. **Find Best Split**: Choose the feature and threshold that best separates classes
3. **Split Data**: Divide data based on the chosen feature/threshold
4. **Repeat Recursively**: Apply the same process to each subset
5. **Stop When**: Reach maximum depth, minimum samples, or pure nodes

#### Splitting Criteria:
- **Gini Impurity**: Measures how often a randomly chosen element would be incorrectly labeled
  - Formula: Gini = 1 - Σ(p_i)²
  - Lower Gini = Better split
- **Entropy**: Measures information gain from a split
  - Formula: Entropy = -Σ(p_i * log₂(p_i))
  - Higher information gain = Better split

## What is a Support Vector Machine (SVM)?

**Support Vector Machines** are powerful supervised learning algorithms used for classification and regression. SVMs find the optimal hyperplane that separates different classes with the maximum margin.

### Key Concepts:

#### The Hyperplane:
- **Hyperplane**: A decision boundary that separates classes
- **Support Vectors**: Data points closest to the hyperplane
- **Margin**: Distance between the hyperplane and nearest support vectors
- **Goal**: Maximize the margin for better generalization

#### How SVM Works:
1. **Find Hyperplane**: Identify the optimal separating hyperplane
2. **Maximize Margin**: Choose hyperplane with largest margin
3. **Support Vectors**: Only support vectors influence the decision boundary
4. **Kernel Trick**: Transform data to higher dimensions for non-linear separation
