## 2. Variables and Data Types

Variables store data values.

Common Python data types:
- int
- float
- string
- boolean

```python
name = 'Alice'
age = 25
height = 5.6
is_student = True

print(type(name))
print(type(age))
print(type(height))
print(type(is_student))
```

## 3. Strings

Strings are sequences of characters.

Common operations:
- Indexing
- Slicing
- Methods

```python
text = 'Python Programming'

print(text[0])
print(text[0:6])
print(text.lower())
print(text.upper())
```

## 4. Operators

Types of operators:
- Arithmetic
- Comparison
- Logical

```python
a = 10
b = 5

print(a + b)
print(a > b)
print(a > 3 and b < 10)
```

## 5. Conditional Statements

```python
age = 18

if age >= 18:
    print('Adult')
else:
    print('Minor')
```

## 6. Loops

Loops allow repeating code.

```python
for i in range(5):
    print(i)
```

```python
x = 0

while x < 5:
    print(x)
    x += 1
```

## 7. Lists

Lists store multiple values.

```python
fruits = ['apple', 'banana', 'mango']

fruits.append('orange')
print(fruits)

fruits.remove('banana')
print(fruits)
```

## 8. Dictionaries

Dictionaries store key-value pairs.

```python
student = {
    'name': 'Ali',
    'age': 22,
    'grade': 'A'
}

print(student['name'])
```

## 9. Functions

Functions allow reusable code.

```python
def add(a, b):
    return a + b

print(add(3, 4))
```
