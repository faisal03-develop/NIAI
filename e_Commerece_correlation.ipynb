```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/faisa/Downloads/train.csv")
df.head()
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.head()
```

```python
corr_matrix = df_encoded.corr()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Matrix")
plt.show()
```

```python
target_corr = corr_matrix["units_sold"].sort_values(ascending=False)
print(target_corr)
```

```python
target_corr.drop("units_sold").plot(kind="bar", figsize=(10,5))
plt.title("Feature Correlation with Units Sold")
plt.show()
```
